### Analyse and visualize sentiment patterns in social media data to understand public opinion and attitudes towards specific topics or brands.

```python
import numpy as np
import pandas as pd
data=pd.read_excel("task4.xlsx")
data.head(11)
```

```python
df = data.loc[:, ~data.columns.str.contains('^Unnamed')]

pd.DataFrame(df)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter positive and negative sentiment data
positive_df = df[df['Sentiment'] == 'Positive']
negative_df = df[df['Sentiment'] == 'Negative']

# Extract data for plotting
positive_sentiment_counts = positive_df.shape[0]  # Count of positive sentiment rows
negative_sentiment_counts = negative_df.shape[0]  # Count of negative sentiment rows

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(['Positive', 'Negative'], [positive_sentiment_counts, negative_sentiment_counts])  # Explicit labels and counts
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Positive vs. Negative Sentiment')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud


# Plot Sentiment Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Sentiment')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()

# Plot Topic Sentiment Count
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='TOPIC', hue='Sentiment')
plt.title('Topic Sentiment Count')
plt.xlabel('Topic')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.show()


# Optional: If you have a time column, you can plot sentiment over time
# Assuming you have a 'Time' column
# df_cleaned['Time'] = pd.to_datetime(df_cleaned['Time'])
# df_cleaned.set_index('Time', inplace=True)



# Bar Plot for Topics
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='TOPIC')
plt.title('Count of Tweets per Topic')
plt.xlabel('Topic')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Ensure the Tweet column is string type and fill NaNs with empty strings
df['Tweet'] = df['Tweet'].astype(str).fillna('')

# Generate Word Cloud for each Sentiment
for sentiment in df['Sentiment'].unique():
    # Filter the DataFrame for the current sentiment
    sentiment_df = df[df['Sentiment'] == sentiment]
    
    # Join all tweets into a single string
    sentiment_text = " ".join(sentiment_df['Tweet'].dropna().astype(str))
    
    # Generate the word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(sentiment_text)
    
    # Display the word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud for {sentiment} Sentiment')
    plt.show()
```
